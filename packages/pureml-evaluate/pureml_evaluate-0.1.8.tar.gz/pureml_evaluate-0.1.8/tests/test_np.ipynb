```python
import joblib
y_true = joblib.load("y_test.pkl").values
y_pred = joblib.load("y_pred.pkl")
sensitive_features = joblib.load("sensitive_features.pkl").to_numpy()
```

```python
sensitive_features
```

```python
print(sensitive_features.ndim, sensitive_features.shape)
```

```python
sensitive_features = sensitive_features.reshape(-1, 1)
print(sensitive_features.ndim, sensitive_features.shape)
sensitive_features
```

```python
import random
import numpy as np

values = ['Indian', 'African', 'American']

list_length = sensitive_features.shape[0]
full_list = values * (list_length // len(values))
full_list += values[:list_length % len(values)]
random.shuffle(full_list)

full_list = np.array(full_list)
print(full_list.shape)
# Print the resulting list
# print(full_list)
```

```python
s_feat = np.column_stack((full_list, sensitive_features))
print(s_feat.shape, s_feat.ndim)
```

```python
s_feat[0:10]
```

```python
s_feat[:, 0]
```

```python
from pureml_evaluate.evaluators.evaluator import eval


values = eval(y_true=y_true, 
              y_pred=y_pred, 
              sensitive_features=s_feat, 
              evaluators=['classification', 'fairness'])
values
```

```python
import json



with open("values_np.json", "w",) as out_file:
    json.dump(values, out_file)
```
